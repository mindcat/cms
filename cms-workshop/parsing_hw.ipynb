### In the lesson materials there is a file called `03_Prod.mdout`. This is a file from the Amber molecular dynamics program.

### Your assignment is to parse this file and write a new file containing a list of the total energies (`Etot`). Name your file `etot.txt`.

#### Extra Challenge: also process other values...

```python
def parse(included_val, ):
    tokens = []
    with open('data/03_Prod.mdout') as f:
#         [tokens.append for t in l.split(' ') for l in f]
        for line in f:
            tokens.extend(line.split())
#         print(tokens)
    print('Number tokens parsed: ', len(tokens))
    f_tokens = []
    for t in range(len(tokens)-1):
#         print(t)
        if (tokens[t-1] == '=' or tokens[t+1] == '='):
            f_tokens.append(tokens[t])
#     print(f_tokens)
    final = ""
    for t in range(len(f_tokens)-1):
        if 'NSTEP'.lower() in tokens[t].lower():
            final = final + '\n' + tokens[t] + " = " + tokens[t+2]
        if tokens[t].lower() in included_val:
            final = final + '\n\t' + tokens[t] + " = " + tokens[t+2]
    return final
        
```

```python
result = parse(['etot', 'eelec'])
```

```python
def cli():
    vals = input("Write a subset of values to a file: [e.g. <filename> <value_1> <value_2> <value_n>]\n\t>").split()
#     print(vals)
    filename = vals[0] + ".out"
    vals = vals[1:len(vals)]
    result = parse(vals)
    with open(filename, "w+") as f:
        f.write(result)
    mesg = "The file " + filename + " has been written with the values: "
    mesg.join(vals)
    print(mesg)
    if input("Type 'y' to see the results... ") == 'y':
        print(result)
```
